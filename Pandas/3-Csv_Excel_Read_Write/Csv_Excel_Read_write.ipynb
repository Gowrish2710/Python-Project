```python
import pandas as ps
dataFrame = ps.read_csv('stock_data.csv')
dataFrame
```

```python
#skipRows method used to skip unwanted row from data frame
dataFrame = ps.read_csv('stock_data.csv',skiprows=1) #It will remove 1st row from data frame
dataFrame
#In below case it is removed header row(which means 1st row from top) so the 2nd row is act as header row
```

```python
#using header 
dataFrame = ps.read_csv('stock_data.csv',header=2)
#it will indicate which row should be act as header row
#Greter than rows will display as rows
#equal row will display as header
#less than rows will not display in data frame
dataFrame
```

```python
#header is none
dataFrame = ps.read_csv('stock_data.csv',header=None) #It will add columns automatically
dataFrame
```

```python
#Header names
dataFrame = ps.read_csv('stock_data.csv',header=None,names=['column1','column2','column3','column4','column5']) 
#It will add columns automatically
#in names ,Column name exceeds,it will new column with all vales as Nan
dataFrame
```

```python
#nrows 
dataFrame = ps.read_csv('stock_data.csv',nrows=2)
#It will display mentioned number of rows without considering header row
# It will display 3 rows (1 header row and 2 value row)
dataFrame
```

```python
#na_values
dataFrame = ps.read_csv('stock_data.csv',na_values=['not available','n.a.']) 
#Search each cells and replace NaN based on na_values
dataFrame
```

```python
#na_values in particular column
dataFrame = ps.read_csv('stock_data.csv',na_values={
    'eps':['not available','n.a.']
    }) 
#Search each cells and replace NaN based on na_values in particular column only
dataFrame
```

```python
#copy datafame into new csv file
dataFrame.to_csv('newCopyDfCsv.csv')
#It will also copy index
```

```python
#If we dont want index in new file,we need to use index attribute as false
dataFrame.to_csv('newCopyDfCsvWithoutIndex.csv',index=False)
```

```python
#particular columns
dataFrame.to_csv('NecessaryColumn.csv',index=False,columns=['tickers','eps'])
```

```python
#Remove haeder
dataFrame.to_csv('WithoutHeader.csv',index=False,header=False)
```

```python
##EXCEL
dataFrameExcel = ps.read_excel("stock_data.xlsx","Sheet1")
dataFrameExcel
```

```python
dataFrameExcel = ps.read_excel("stock_data.xlsx","Sheet1",na_values=['n.a.'])
dataFrameExcel
```

```python
#We can use converters for replacement of Nan
def convert_cell(cell):
    if cell == 'n.a.':
        return 'Gowrish'
    return cell
dataFrameExcel = ps.read_excel("stock_data.xlsx","Sheet1",converters={
    'price':convert_cell
})
#When read a excel/csv file,'price' column each cell value it will call convert_cell method
dataFrameExcel
```

```python
#Write Excel file
dataFrameExcel.to_excel('WriteNewExcel.xlsx',sheet_name='Gowrish')
```

```python
#Start row and column
dataFrameExcel.to_excel('startrowcolNewExcel.xlsx',sheet_name='Gowrish',startrow=2,startcol=2)
```

```python
#write values to different sheets using excelwriter
df_stocks = ps.DataFrame({
    'tickers': ['GOOGL', 'WMT', 'MSFT'],
    'price': [845, 65, 64 ],
    'pe': [30.37, 14.26, 30.97],
    'eps': [27.82, 4.61, 2.12]
})
df_weather =  ps.DataFrame({
    'day': ['1/1/2017','1/2/2017','1/3/2017'],
    'temperature': [32,35,28],
    'event': ['Rain', 'Sunny', 'Snow']
})
```

```python
with ps.ExcelWriter('ExcelWriter.xlsx') as writer:
    df_stocks.to_excel(writer,sheet_name='sheetOne')
    df_weather.to_excel(writer,sheet_name='sheetTwo')
```
