```python
#Replace method
import pandas as ps
import numpy as py
dataFrameCsv = ps.read_csv('weather_data.csv')
dataFrameCsv
```

```python
#Using replace method ,we can replace cell value
New_Replace = dataFrameCsv.replace(-99999,py.NaN)
New_Replace
```

```python
#multiple value replace by one
New_Replace = dataFrameCsv.replace([-99999,7,'0'],py.NaN)
New_Replace
```

```python
#multiple value replace by one using dictionary
New_Replace = dataFrameCsv.replace({
    'temperature':-99999,
    'event':'0'
},py.NaN)
New_Replace
```

```python
#multiple value replace by one using dictionary
New_Replace = dataFrameCsv.replace({
    -99999:py.NaN,
    '0':'Sunny Day'
})
New_Replace
```

```python
#Using Regular Expression
New_Replace = dataFrameCsv.replace({
    'event' : '[A-Za-z]'
},'',regex=True)
New_Replace
```
