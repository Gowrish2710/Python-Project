```python
import pandas as ps
dataFrame = ps.read_csv('aapl_no_dates.csv')
dataFrame.head()
```

```python
ranges = ps.date_range(start='1/1/2017',end='1/31/2017',freq='B',) #B-Business days
ranges
```

```python
dataFrame.set_index(ranges,inplace=True)
dataFrame
```

```python
%matplotlib inline
dataFrame.Close.plot()
```

```python
dataFrame['2017-01-01':'2017-01-30'].Close.mean()
```

```python
#It will also add weekend values from previous day value
dataFrame.asfreq('D',method='pad')
```

```python
#Based on Hour
dataFrame.asfreq('H',method='pad')
```
