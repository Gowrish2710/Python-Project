```python
#Handle missing Data Using Fillna,dropna and interpolate methods
```

```python
import pandas as ps
dataFrameCsv = ps.read_csv('weather_data.csv')
dataFrameCsv
```

```python
#Parse date
dataFrameCsv = ps.read_csv('weather_data.csv',parse_dates=['day'])
dataFrameCsv.set_index('day',inplace=True)
dataFrameCsv
```

```python
#fillna method
fillnaCsv = dataFrameCsv.fillna(0) #it replace Nan with 0
fillnaCsv
```

```python
#fillna with particular column
fillnaCsv = dataFrameCsv.fillna({
    'temperature':0,
    'event':'Climate'
})
fillnaCsv
```

```python
#forward fill
fillnaCsv = dataFrameCsv.fillna(method='ffill')#It will copy the current value to immediat next Nan Cell
fillnaCsv
```

```python
#forward fill
dataFrameFfill = ps.read_csv('weather_data_ForawardFill.csv')
fillnaCsv = dataFrameFfill.fillna(method='ffill')
#if initial row has Nan value means,it will not change
fillnaCsv
```

```python
#backward fill
fillnaCsv = dataFrameCsv.fillna(method='bfill')#It will copy the current value to immediate previous Nan Cell
fillnaCsv
```

```python
#ffill with limit
fillnaCsv = dataFrameCsv.fillna(method='ffill',limit = 1)
#Maximum limit of copying from to another cell is one
#If continously Nan cell comes,it will fill only one cell ,others remain as Nan
fillnaCsv
```

```python
#interpolate
interPolCsv = dataFrameCsv.interpolate()
#It will fill value with difference based on inbetween cells and provide appromiate guessable value to cell
#It alaways work with numbers
interPolCsv
```

```python
interPolCsv=dataFrameCsv.interpolate(method='time')
interPolCsv
#here it will fill guessable value using time (i.e., day column)
```

```python
#dropna
dropCsv = dataFrameCsv.dropna()
#It will remove row which has atleaset a one Nan value in cell
dropCsv
```

```python
#dropna with thresh
#No of valid values should be there for avoid drop
dropCsv = dataFrameCsv.dropna(thresh=2)
dropCsv
```
